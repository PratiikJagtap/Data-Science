```python
# create a new column

data['DS'] = 'Data Science'
```

```python
# Delete the column
data.drop(['DS'], axis=1, inplace=True) # axis=1 refer to column
```

```python
# Delete the row
data.drop(1, axis=0, inplace=True) # axis=0 refer to row
```

```python
# deleting multiple rows
data.drop([3,5], axis=0)
```

```python
# Copy of file for backup reasons
new_data = data.copy()
```

```python
# creating new column using existing column
data['Family'] = data['SibSp'] + data['Parch']
```

```python
# aranging 
data.sort_values('Pclass', ascending=True, ignore_index=True) # ignore_index it let never change the index numbers of table
```

```python
data.sort_values(by=['Age','Pclass'], ascending=[0,1]) # used for multiple columns ascending values
# 0 is treated as False and 1 as True
```
