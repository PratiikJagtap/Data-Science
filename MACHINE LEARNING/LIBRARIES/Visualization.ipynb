# Introduction to MatplotLib

- Using pandas, we can get insights of data by plotting the graph
- We use matplotlib (matlab style plotting) and seaborn (seaborn is built on matplotlib)
- For categorical data, we use bar plot, count plot, pie plot. (Univariate analysis)
- For numerical data, we use histogram or distribution plot (Univariate analysis)
- In bivariate analysis, if x and y are continuous we use scatter plot
- In bivariate analysis, if x or y is discrete and other is continuous, we use bar plot

# Pie Plot

```python
plt.style.use("ggplot")

labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
explode = [0.1, 0, 0, 0]

plt.title("Career Runs of Indian Batsmen")
plt.pie(
    runs,
    labels=labels,
    explode=explode, # Pull one slice out of cirlce
    colors=colors,
    # wedgeprops={'edgecolor': 'black', 'linewidth': 2, 'linestyle': '--'}
    shadow=True,
    startangle=340, # To rotate the pie plot
    autopct='%1.1f%%' # To show percentage 
    
)
plt.axis('equal')
plt.show()
```

# stack plot

```python
days = [1, 2, 3, 4, 5, 6, 7]  # Days of the week
studying = [3, 4, 3, 5, 4, 3, 4]
playing = [2, 2, 1, 1, 2, 3, 2]
watching_tv = [2, 1, 2, 2, 1, 1, 1]
sleeping = [5, 5, 6, 5, 6, 5, 5]

labels = ['Studying', 'Playing', 'Watching TV', 'Sleeping']
colors = ['skyblue', 'lightgreen', 'gold', 'lightcoral']

plt.figure(figsize=(10,6)) # size of the plot
plt.stackplot(days, studying, playing, watching_tv, sleeping,
              labels=labels, colors=colors, alpha=0.8) # alpha shows the ocupacity

plt.legend(loc='upper left')  # Location of the legend
plt.title('Weekly Activity Tracker')
plt.xlabel('Day')
plt.ylabel('Hours')
plt.grid(True)
plt.show()
```

# Scatter plot

```python
plt.style.use('classic')
study_hours = [1,2,3,4,5,6,7,8,9]
exam_scores = [40, 45, 90, 55, 60, 25, 75, 85, 90]
colors =  ['red' if score<50 else 'green' for score in exam_scores]
sizes = [score * 2 for score in exam_scores]
social_scores = [12, 34, 67, 55, 67, 145, 56, 999, 445] #  This numbers only affect colors
plt.scatter(study_hours, exam_scores, c=social_scores, s=sizes, cmap='viridis')
plt.colorbar(label='Score')
# add labels to dots
for i in range(len(study_hours)):
    plt.annotate(f'S{i+1}', (study_hours[i] + 0.25, exam_scores[i]))
plt.grid(True)
plt.xlabel('Study Hours')
plt.ylabel('Exam Scores')
plt.title('Study vs Marks')
plt.show()
```

```python
# Multiple Groups in One Plot

# Assume two groups: Class A and Class B
class_a_hours = [2, 4, 6, 8]
class_a_scores = [45, 55, 65, 85]
 
class_b_hours = [1, 3, 5, 7, 9]
class_b_scores = [40, 50, 60, 70, 90]

plt.scatter(class_a_hours, class_a_scores, label='A', color='Blue')
plt.scatter(class_b_hours, class_b_scores, label='B', color='red')
plt.title('Scatter Plot: Class A vs Class B')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.legend()
plt.grid(True)
plt.show()
```

# SubPlot

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y1 = [i * 2 for i in x]
y2 = [i ** 2 for i in x]

# Create a figure with 1 row and 2 columns
plt.subplot(1, 2, 1)  # (rows, cols, plot_no)
plt.plot(x, y1)
plt.title('Double of x')

plt.subplot(1, 2, 2)
plt.plot(x, y2)
plt.title('Square of x')

plt.tight_layout()
plt.show()
```

```python
# Using plt.subplots() for Clean Code
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

axs[0].plot(x, y1)
axs[0].set_title('x * 2')

axs[1].plot(x, y2)
axs[1].set_title('x squared')
 
fig.suptitle('Simple Comparison Plots', fontsize=14)
fig.tight_layout()
fig.subplots_adjust(top=0.85)  # So title doesn't overlap
# fig.savefig('my_plots.png')    # Save as image

plt.show()
```

```python
# Looping Over Subplots
x = np.linspace(0, 10, 100)
y1 = x * 2
y2 = x ** 2
y3 = np.sqrt(x)
y4 = 10 - x

fig, axs = plt.subplots(2, 2, figsize=(8, 6))
ys = [y1, y2, y3, y4]
titles = ['x * 2', 'x squared', 'sqrt(x)', '10 - x']

for i in range(2):
    for j in range(2):
        idx = i * 2 + j
        axs[i, j].plot(x, ys[idx])
        axs[i, j].set_title(titles[idx])

plt.tight_layout()
plt.show()
```

# Introduction to Seaborn

#### Seaborn is a Python library built on top of Matplotlib that makes it easier and prettier to create complex, beautiful visualizations.

### Why Seaborn?
- Matplotlib is powerful but very low-level.
- Seaborn adds high-level features like automatic styling, themes, color palettes, and dataframe integration.
- Seaborn comes with built in Datasets
- Makes complex plots (like boxplots, violin plots, heatmaps, pairplots) very easy.

#### In short:
- Less code
- Better-looking graphs
- Easy handling of DataFrames (like from pandas)


```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df = sns.load_dataset('iris') # we can directly load data by using sns library
```

```python
# Line Plot
sns.lineplot(x='sepal_length', y='sepal_width', hue='species', data=df, ci=0) #hue means legend
plt.show()
```

```python
#Scatter Plot
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=df, palette='pastel') #hue means legend
plt.title('Scatter plot')
plt.show()
```

```python
# bar plot

sns.barplot(x='species', y='sepal_length', data=df, ci=0, palette='Set3')
plt.show()
```

```python
# Count Plot
sns.countplot(x='species', data=df) # this counts values
plt.show()
```

```python
# Box Plot

sns.boxplot(x='sepal_width', data=df)
plt.show()
```

```python
# Box plot according to categories

sns.boxplot(x='sepal_width', y='species', data=df)
plt.show()
```

```python
# Violinplot
sns.violinplot(x='species', y='sepal_width', data=df, palette='Set3')
plt.show()
```

```python
# Stripplot
#It is basically creates a scatter plot based on category
sns.stripplot(x="species", y="sepal_width", data=df, jitter=True)  # jitter=True spreads the points
plt.show()
```

```python
# Histogram
# A histogram is basically used to represent data provided in a form of some groups.
sns.histplot(x='petal_width', data=df, color='red')
plt.show()
```

```python
sns.histplot(x='petal_width', data=df, hue='species', kde=True)
plt.show()
```

```python
# Distplot
sns.distplot(df['petal_width'])
plt.show()
```

```python
# Pairplot
sns.pairplot(data=df)
plt.show()
```

```python
#Categorical plot

sns.catplot(x='sepal_length', y='species', data=df)
plt.show()
```
